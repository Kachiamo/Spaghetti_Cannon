# Step 2: Import the OHLC dataset into Panadas Dataframe.

```python
# Import the OHLCV dataset into a Pandas Dataframe
tick_df = pd.read_csv(
    Path("Resources/META.csv"), 
    index_col='Date', 
    infer_datetime_format=True, 
    parse_dates=True
)

# Review the DataFrame
tick_df.head()
```

```python
# Plot the DataFrame with hvplot
tick_df["Close"].hvplot(ylabel='Price')
```

# Step 3: Generate the Stochastich Oscillator Technical Indicators

```python
# def generate_stochastic_signals(df, high_col='High', low_col='Low', close_col='Close', k_period=14, d_period=3, overbought=80, oversold=20):
    # tick_df = generate_stochastic_signals(tick_df)
k_period=14
d_period=3
overbought=80
oversold=20

# Calculate %K and %D
tick_df['%K'] = 100 * ((tick_df['Close'] - tick_df['Low'].rolling(window=k_period).min()) / (tick_df['High'].rolling(window=k_period).max() - tick_df['Low'].rolling(window=k_period).min()))
tick_df['%D'] = tick_df['%K'].rolling(window=d_period).mean()

# Generate entry/exit signals
tick_df['Long Signal'] = (tick_df['%K'] < oversold) & (tick_df['%D'] < oversold)
tick_df['Short Signal'] = (tick_df['%K'] > overbought) & (tick_df['%D'] > overbought)
```

```python
display(tick_df.head())
display(tick_df.tail())
```

```python
tick_df["Signals"] = np.where(tick_df['Short Signal'], -1,       
np.where(tick_df['Long Signal'], 1,0))

tick_df
```

# Step 4: Plot Trading Indicator with Entry/Exit Points

```python
# Visualize entry position relative to close price
entry = tick_df[tick_df["Signals"] == 1.0]["Close"].hvplot.scatter(
    color='green',
    marker='^',
    size=200,
    legend=False,
    ylabel='Price in $',
    width=1000,
    height=400
)

# Visualize exit position relative to close price
exit = tick_df[tick_df["Signals"] == -1.0]["Close"].hvplot.scatter(
    color='red',
    marker='v',
    size=200,
    legend=False,
    ylabel='Price in $',
    width=1000,
    height=400
)

# # Visualize close price for the investment
security_close = tick_df["Close"].hvplot(
    line_color='lightgray',
    ylabel='Price in $',
    width=1000,
    height=400
)

# Overlay plots
entry_exit_plot = entry * exit * security_close
entry_exit_plot
```
